```python
from qiskit import QuantumCircuit
import numpy as np

qc = QuantumCircuit(3)

qc.x([0,2])

qc.h(0)
qc.cx(0,1)
qc.rx(np.pi/2,2)
# qc.h(1)
qc.y(1)
qc.y(1)
# qc.h(1)
qc.rx(np.pi/2,2)
# qc.rx(pi)
qc.cx(0,1)
qc.h(0)

qc.measure_all()

qc.draw(reverse_bits=False)
```

```python
from qiskit_ibm_runtime import SamplerV2
from qiskit_aer import AerSimulator

sampler_aer = SamplerV2(backend=AerSimulator())

job = sampler_aer.run([(qc, None, 100)])
pub_result = job.result()[0]
print(f"1st Circuit Counts: \n {pub_result.data.meas.get_counts(0)} \n")
```

```python
from qiskit import transpile
from qiskit.providers.basic_provider import BasicSimulator
from qiskit_aer import AerSimulator


simulator_basic = BasicSimulator()
simulator_aer = AerSimulator()
```

```python
# qc = QuantumCircuit(2,2)
# qc.h(0)

qc = QuantumCircuit(3)

qc.x([0,1])

# qc.h(0)
qc.cx(0,1)
# qc.rx(np.pi/2,2)
# qc.h(1)
# qc.y(1)

qc.save_statevector()

qc.draw()
```

```python
qc_aer = transpile(qc, backend=simulator_aer)

result = simulator_aer.run(qc_aer).result()
psi = result.get_statevector()
psi
```
