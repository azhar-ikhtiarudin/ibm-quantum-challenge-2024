```python
from qiskit import QuantumCircuit
import numpy as np

qc = QuantumCircuit(3)

qc.x([0,2])

qc.h(0)
qc.cx(0,1)
qc.rx(np.pi/2,2)
# qc.h(1)
qc.y(1)
qc.y(1)
# qc.h(1)
qc.rx(np.pi/2,2)
# qc.rx(pi)
qc.cx(0,1)
qc.h(0)

qc.measure_all()

qc.draw(reverse_bits=False)
```

```python
from qiskit_ibm_runtime import SamplerV2
from qiskit_aer import AerSimulator

sampler_aer = SamplerV2(backend=AerSimulator())

job = sampler_aer.run([(qc, None, 100)])
pub_result = job.result()[0]
print(f"1st Circuit Counts: \n {pub_result.data.meas.get_counts(0)} \n")
```

```python
import numpy as np

# print(np.cos(np.pi))

# theta = 7*np.pi/11
theta_1 = np.pi/8
theta_2 = np.pi/4

c2 = np.cos(2*theta_1)
s2 = np.sin(2*theta_1)
Ref_1 = np.array([[c2, s2],
                [s2, -c2]]) # ref pi/8

c2 = np.cos(2*theta_2)
s2 = np.sin(2*theta_2)
Ref_2 = np.array([[c2, s2],
                [s2, -c2]]) # ref pi/4

state = np.array([[1],[0]])

result_a = Ref_2 @ Ref_1 @ state
result_b = Ref_1 @ Ref_2 @ state

# result = Ref @ R @ Ref @ R @ Ref @ R @ Ref @ R @ state 
print(result_a)
print(result_b)
# print(result[0]**2)
# print(result[1]**2)
# print(R @ state)

import math

def dotproduct(v1, v2):
  return sum((a*b) for a, b in zip(v1, v2))

def length(v):
  return math.sqrt(dotproduct(v, v))

def angle(v1, v2):
  return math.acos(dotproduct(v1, v2) / (length(v1) * length(v2)))

print(angle(result_a, result_b))
```
